```python
%matplotlib inline
#from string import letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pf = pd.read_csv('E:\ML\matches.csv', index_col=0, encoding='latin-1')
```

```python
pf.head()
```

```python
sns.set_style('whitegrid')
r = sns.countplot(x="city", data=pf)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.axes().set_title('Top Venues')
r.set(ylabel='Number of Matches', xlabel='City')
r.figure.savefig('top_venues.png')
```

```python
kf=pf[(pf.team1 != 'Pune Warriors')&(pf.team1!='Gujarat Lions') & (pf.team1!='Rising Pune Supergiants') & (pf.team1!='Kochi Tuskers Kerala')]
kf=kf[(pf.team2 != 'Pune Warriors')&(pf.team2!='Gujarat Lions') & (pf.team2!='Rising Pune Supergiants') & (pf.team2!='Kochi Tuskers Kerala')]
```

```python
len(kf)
```

```python
sns.set_style('whitegrid')
t = sns.countplot(x="toss_winner", hue="toss_decision", data=kf)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.axes().set_title('Toss Decisions by Team')
t.set(ylabel='Number of Matches', xlabel='Team')
t.figure.savefig('toss_decision.png')
```

```python
sns.set_style('whitegrid')
q = sns.swarmplot(x='winner', y='win_by_wickets',data=kf[(kf.win_by_wickets!=0)])
plt.xticks(rotation = 90)
plt.ylim(0,None)
plt.tight_layout()
plt.axes().set_title('Win By Wickets')
q.set(ylabel='Number of Wickets', xlabel='Team')
q.figure.savefig('win_wicket.png')
```

```python
sns.set_style('whitegrid')
p = sns.swarmplot(x='winner', y='win_by_runs',data=kf[(kf.win_by_runs!=0)])
plt.xticks(rotation = 90)
plt.ylim(0,None)
plt.tight_layout()
p.set(ylabel='Win By Runs', xlabel='Team')
p.figure.savefig('win_run.png')
```

```python
sns.set_style('whitegrid')
t = sns.countplot(x="winner", data=kf)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.axes().set_title('Number of Wins')
t.set(ylabel='Number of Matches', xlabel='Team')
t.figure.savefig('number_win.png')
```

```python
sns.set_style('whitegrid')
sns.color_palette("cubehelix", 8)
t = sns.countplot(x="toss_winner", data=kf)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.axes().set_title('Number of Toss Wins')
t.set(ylabel='Number of Matches', xlabel='Team')
t.figure.savefig('num_toss_wins.png')
```

```python
sns.set_style('whitegrid')
sns.color_palette("cubehelix", 8)
t = sns.countplot(x="player_of_match", data=kf, order=kf.player_of_match.value_counts().iloc[:15].index)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.axes().set_title('Man of the Match')
t.set(ylabel='Number of Awards', xlabel="Player's Name")
t.figure.savefig('player_of_the_match.png')
```
